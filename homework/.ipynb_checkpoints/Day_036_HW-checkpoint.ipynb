```python
#z score test
def outliers_z_score(ys, times):
    meany = np.mean(ys)
    sdy = np.std(ys)
    z_scores = [(y - meany) / sdy for y in ys]
    return np.where(np.abs(z_scores) > times)
out1 = outliers_z_score(df_train["Fare"], 3)
out1[0]
df_train["Fare"][out1[0]]
#z分數因為是由平均及標準差計算，容易受極端值影響，因此範圍較大，所以列為極端值的數字較少
```

```python
print(type(out1[0]), type(out1))
```

```python
#IQR test
def outliers_iqr(ys):
    quartile_1, quartile_3 = np.nanpercentile(ys, [25, 75])
    iqr = quartile_3 - quartile_1
    return np.where((ys > quartile_3+1.5*iqr) | (ys < quartile_1-1.5*iqr))
out2 = outliers_iqr(df_train["Fare"])
df_train["Fare"][out2[0]]
#IQR因不受極端值影響，列為極端值的數字較多
```
