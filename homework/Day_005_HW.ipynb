```python
import numpy as np
english_score = np.array([55,89,76,65,48,70])
math_score = np.array([60,85,60,68,np.nan,60])
chinese_score = np.array([65,90,82,72,66,77])
```

```python
# 請計算各科成績平均、最大值、最小值、標準差，其中數學缺一筆資料可忽略？
print("max:", np.max(english_score), "min:", np.min(english_score),
      "mean:", np.mean(english_score), "sd:", np.std(english_score))
print("max:", np.nanmax(math_score), "min:", np.nanmin(math_score),
      "mean:", np.nanmean(math_score), "sd:", np.nanstd(math_score))
print("max:", np.max(chinese_score), "min:", np.min(chinese_score),
      "mean:", np.mean(chinese_score), "sd:", np.std(chinese_score))
```

```python
# 第五位同學補考數學後成績為 55，請計算補考後數學成績平均、最大值、最小值、標準差？
math_score[4] = 55
math_score
```

```python
print("max:", np.nanmax(math_score), "min:", np.nanmin(math_score),
      "mean:", np.nanmean(math_score), "sd:", np.nanstd(math_score))
```

```python
# 用補考後資料找出與國文成績相關係數最高的學科？
arr = np.array([math_score, chinese_score, english_score])
np.corrcoef(arr)
```
