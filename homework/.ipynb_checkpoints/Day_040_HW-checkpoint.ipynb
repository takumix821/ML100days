## 作業
在鐵達尼資料集中，
透過進階補值方法 KNN，將鐵達尼號中的 Titanic_train
.csv 中的 age 的遺失值進行補值。    
* step1: 觀察 Age 和 Pclass 與 Sex 是否有關連性?
* step2: 如果有關連性，運用 KNN ，取出 Age 、 Pclass、 Sex的資料，以 Sex 與 Pclass 補 Age 遺失值。 

```python
# import library
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
import math
import statistics
import seaborn as sns
from IPython.display import display
```

## 讀入資料

```python
df = pd.read_csv('Titanic_train.csv')
df.info()
```

```python
df.describe()
```

## 觀察 Age

```python
complete_data = df[['Age','Pclass','Sex']].dropna()
Sex_mapping = {'male' : 1, 'female' : 0}
complete_data['Sex'] = complete_data['Sex'].map(Sex_mapping)
complete_data.head(10)
```

```python
# step1: 觀察 Age 和 Pclass 與 Sex 是否有關連性?
def corrtest(corr):
    if corr >= 0.7: print('高度正相關')
    elif (corr >= 0.3) & (corr < 0.7): print('中度正相關')
    elif (corr < 0.3) & (corr > 0): print('低度正相關')
    elif (corr > -0.3) & (corr < 0): print('低度負相關')
    elif (corr <= -0.3) & (corr > -0.7): print('中度負相關')
    elif corr <= -0.7: print('高度負相關')
    elif corr == 0: print('零相關')
```

```python
corrPclass, _ = stats.pearsonr(complete_data['Age'], complete_data['Pclass'])
print('Pclass:', corrPclass, corrtest(corrPclass))
corrSex, _ = stats.pearsonr(complete_data['Age'], complete_data['Sex'])
print('Sex:' ,corrSex, corrtest(corrSex))
```

```python
# step2: 如果有關連性，運用 KNN ，取出 Age 、 Pclass、 Sex的資料，以 Sex 與 Pclass 補 Age 遺失值。
df['Sex'] = df['Sex'].map(Sex_mapping)
from sklearn import preprocessing
from sklearn.metrics.pairwise import nan_euclidean_distances
display(pd.DataFrame(nan_euclidean_distances(df[['Age','Pclass','Sex']])))
```

```python
value_neighbors = 1
from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors = value_neighbors)
df_filled = pd.DataFrame(imputer.fit_transform(df[['Age','Pclass','Sex']]))
```
